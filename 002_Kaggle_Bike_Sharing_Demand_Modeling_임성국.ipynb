  * registered : 사전 등록된 사용자 대여 횟수
  * count : 총 대여 횟수 -> y

# Data Load
* 'Bike_Sharing_Demand.csv' Github에서 읽어오기

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/rusita-ai/pyData/master/Bike_Sharing_Demand.csv'
DF = pd.read_csv(url)

DF.info()
```

```python
DF.head()
```

# I. Data Preprocessing

> ## 1) 'String' -> 'Date'

* 'year', 'month', 'day', 'hour' 추출

```python
DF['datetime'] = DF.datetime.apply(pd.to_datetime)

DF['year'] = DF.datetime.apply(lambda x : x.year)
DF['month'] = DF.datetime.apply(lambda x : x.month)
DF['day'] = DF.datetime.apply(lambda x : x.day)
DF['hour'] = DF.datetime.apply(lambda x: x.hour)
```

```python
DF.head()
```

> ## 2) Drop Columns

* 'datetime', 'casual', 'registered'

```python
drop_columns = ['datetime', 'casual', 'registered']

DF.drop(drop_columns, axis = 1,inplace = True)
```

```python
DF['atemp2']=DF['atemp']**2
```

```python
DF['windspeed2']=DF['windspeed']**2
```

```python
DF['season2']=DF['season']**2
DF['month2']=DF['month']**2
DF['hour2']=DF['hour']**2
DF['hour3']=DF['hour']**3
DF['hour4']=DF['hour']**4
```

# Modeling

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(DF,
                                       test_size = 0.3,
                                       random_state = 2045)

train_set.shape, test_set.shape
```

## (1) count ~ atemp+windspeed

```python
import statsmodels.formula.api as smf

Model_1 = smf.ols(formula = 'count ~ atemp2+atemp+windspeed2+windspeed',
                  data = train_set).fit()
```

```python
y_hat_1 = Model_1.predict(test_set[['atemp2','atemp','windspeed2','windspeed']])
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
mse1 = mean_squared_error(test_set['count'], y_hat_1)

np.sqrt(mse1)
```

## (2) count ~ (1) + season+month+hour

```python
import statsmodels.formula.api as smf

Model_2 = smf.ols(formula = 'count ~ atemp2+atemp+windspeed2+windspeed+season2+season+month2+month+hour4+hour3+hour2+hour',
                  data = train_set).fit()
```

```python
y_hat_2 = Model_2.predict(test_set[['atemp2','atemp','windspeed2','windspeed','season2','season','month2','month','hour4','hour3','hour2','hour']])
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
mse2 = mean_squared_error(test_set['count'], y_hat_2)

np.sqrt(mse2)
```

## (3) count ~ (2) + holyday+year+weather

```python
import statsmodels.formula.api as smf

Model_3 = smf.ols(formula = 'count ~ atemp2+atemp+windspeed2+windspeed+season2+season+month2+month+hour4+hour3+hour2+hour+holiday+year+weather',
                  data = train_set).fit()
```

```python
y_hat_3 = Model_3.predict(test_set[['atemp2','atemp','windspeed2','windspeed','season2','season','month2','month','hour4','hour3','hour2','hour','holiday','year','weather']])
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
mse3 = mean_squared_error(test_set['count'], y_hat_3)

np.sqrt(mse3)
```
